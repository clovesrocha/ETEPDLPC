```python
# ETE PORTO DIGITAL
# DISCIPLINA: LPC
# PROF.: CLOVES ROCHA

# EXEMPLO 1: 
# CONSTRUIR UMA LISTA E IMPRIMIR SEUS ELEMENTOS

from __future__ import print_function

n = 100;

# Definir vetor/lista com: (100, 110, 120, 130,...100+n-1)
vetor = []; #define um vetor vazio
#for = para
for i in range(0, n, 1): # ou mais simples neste caso "for i in range(n)"
  vetor.append(100+i); # anexe mais um elemento à lista

# Imprime o vetor/lista de uma só vez (na mesma linha)
print('\n O vetor: ', end=''); # informe o que vira impresso a seguir (na mesma linha devido ao parametro end='')
print(vetor);

# Imprimir vetor em linha única na forma "(100, 100) (110, 110)..."
print('\n Novamente o vetor impresso de 2 modos: '); # informe o que virá impresso a seguir (e "quebre" a linha)
i = 0;
for item in vetor : # item in vetor é um iterador, item começa em vetor[0], depois vetor [1] e assim por diante
  print("(%2d, %2d) " % (item, vetor[i]), end='');
  i += 1;
  print(); # quebre a linha


        




```
